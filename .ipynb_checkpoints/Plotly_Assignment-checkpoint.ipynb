## Q1: Combine the two datsets as one dataframe

```python
data = pd.merge(movies, credits)
data 
```

## Q2: Remove the following columns:
- 'homepage','tagline','id','overview','status','original_title','movie_id'.

```python
data.drop(columns =['homepage','tagline','id','overview','status','original_title','movie_id'], inplace = True)
data.shape 
```

```python
movies.head(3)
```

## Q3: Create Genres dataframe that hase two coulumns Genres and Frequency
- Hint: Use Counter to calculate Frequency
- Add new catagory to Genres called Others when the Frequency < 200

```python
x = pd.DataFrame(columns = ['Genres'])
from ast import literal_eval

x['Genres'] = data['genres'].apply(literal_eval)

print(x['Genres'][0])
print('--------------')
print(x['Genres'][0][0])
print('--------------')
print(x['Genres'][0][0].get("id"))
```

```python
temp = pd.DataFrame(columns = ['name'])
for i in x['Genres']: 
    for j in i:
        temp.loc[len(temp)] = [j.get("name") ]
```

```python
temp
```

```python
Genres = pd.DataFrame(columns = ['Genres', 'Frequency'])

Genres['Genres'] = temp['name'].value_counts().index
Genres['Frequency'] = temp['name'].value_counts().values

for i in range (len(Genres)):
    if Genres['Frequency'][i] < 200:
        Genres['Genres'][i] = 'Others'
Genres
```

## Q4: Use bar chart draws the relationship between genres of movies and Frequency.


```python
px.bar(data_frame=Genres, x='Genres', y='Frequency', color='Genres')
```

## Q5: Use pie chart to draw top 5 languages.
- Hint: use counter to create dataframe contains production_countries with Frequencies.


```python
x = pd.DataFrame(columns = ['production_countries'])
from ast import literal_eval

x['production_countries'] = data['production_countries'].apply(literal_eval)

temp2 = pd.DataFrame(columns = ['lang'])
for i in x['production_countries']: 
    for j in i:
        temp2.loc[len(temp2)] = [j.get("iso_3166_1") ]
        
langF = pd.DataFrame(columns = ['production_countries', 'Frequency'])

langF['production_countries'] = temp2['lang'].value_counts().index
langF['Frequency'] = temp2['lang'].value_counts().values
langF[:5]
```

```python
x = langF[:5]
px.pie(data_frame=x, names='production_countries', values='Frequency')
```

## Q6: Use WordCloud to draw genres.

```python
text = ''
for i in Genres['Genres'][ Genres['Genres'] != 'Others']:
    text = text + " "+i

plt.imshow(WordCloud(background_color='white').generate(text))
```
